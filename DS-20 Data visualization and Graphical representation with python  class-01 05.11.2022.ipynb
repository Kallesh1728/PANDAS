Histogram 

A Histogram takes in a series of data and devides the data into a number of bins. it then plots the frequency data points in each bin(i.e the interval of points). It is useful in understanding the count of data ranges. when to use : we should use histogram ehen we need the count of the variables in aplot

```python
#hire it is providing the bins of 20 20 intervals, in case we have nat provided the bins bins will take default value will bw 10-10 intervals 
a=np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=[0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```
